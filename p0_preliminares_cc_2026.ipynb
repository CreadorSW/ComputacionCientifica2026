## Diferencia en sintaxis.

Cada lenguaje posee su sintaxis. Pior ejemplo, en Octave, sería normal escribir `a=127;b=13;q=0;`, pero en Python sería un error por lo siguiente:

- Python está diseñado para leerse como prosa. El uso de ; amontona el código y dificulta la detección de errores a simple vista.

- Estilo (PEP 8): La guía de estilo oficial de Python recomienda usar saltos de línea para separar instrucciones.

- Innecesario al final: A diferencia de lenguajes como C++ o Java, poner un ; al final de una línea solitaria no tiene ningún efecto y se considera redundante.

¿Cómo lo haríamos entonces? La forma usual sería

```python
a = 127
b = 13
q = 0
```
O utilizar la *asignación múltiple* `a, b, q = 127, 13, 0`.



Dicho esto, pasemos a la acción.
